```python
from Database import Database
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
db_file = "sqlite:///../data/articles.db"
```

```python
db = Database(db_file)
```

```python
results = db.get_tesla_mentions()
```

```python
print(results[0])
```

```python
dates = [result[0] for result in results]

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(dates, bins=30, color='blue', edgecolor='black')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.title('Histogram of Tesla Mentions Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
